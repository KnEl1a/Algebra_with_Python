```python
import numpy as np
import matplotlib.pyplot as plt

def graficar_r2(vectores):
    # Convertir los vectores a tipo de datos float64 para evitar problemas de tipos
    vectores = [vector.astype(np.float64) for vector in vectores]

    vectores_lista = []  # Lista para almacenar los vectores convertidos
    suma = np.zeros_like(vectores[0])  # Inicializar la suma como un vector de ceros del mismo tamaño que los vectores

    for i, vector in enumerate(vectores):
        c = len(vector)
        print(f"\nEl vector({i+1}) = {vector} tiene {c} componentes")
        try:
            if c != 2:
                raise ValueError("Error, todos los vectores deben tener exactamente 2 componentes")
        except ValueError as e:
            print(e)
            return

        vectores_lista.append(vector.tolist())

    for vector in vectores:
        suma += vector

    # Calcular los límites automáticamente
    x_min = min(0, *np.min(vectores, axis=0), np.min(suma))
    x_max = max(0, *np.max(vectores, axis=0), np.max(suma))
    y_min = min(0, *np.min(vectores, axis=0), np.min(suma))
    y_max = max(0, *np.max(vectores, axis=0), np.max(suma))

    """
    usamos '*' para ...
    El uso de * en Python se conoce como el operador de desempaquetado (unpacking operator).
    Cuando se utiliza junto con una lista o un arreglo, desempaqueta los elementos de la lista o el arreglo y
    los pasa como argumentos separados a una función.
    """

    # Graficar los vectores
    for idx, vector in enumerate(vectores):
        plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='C'+str(idx), label=f'Vector {idx+1}')

    #plt.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1, color='g', label='Suma')
    plt.xlim(x_min-1, x_max+1)  # Ampliar los límites para mejor visualización
    plt.ylim(y_min-1, y_max+1)  # Ampliar los límites para mejor visualización
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()

    print("La suma de los vectores es:", suma)
    print("El producto punto de los vectores es:", np.dot(vectores[0], vectores[1]))

# Ejemplo de uso
vector1 = np.array([1, 2])
vector2 = np.array([3, -1])
vector3 = np.array([-2, 4])

graficar_r2([vector1, vector2, vector3])
```

```python
#grafica en r3

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def graficar_r3(vectores):
    suma = np.zeros_like(vectores[0])  # Inicializar la suma como un vector de ceros del mismo tamaño que los vectores

    # Convertir los vectores a tipo de datos float64 para evitar problemas de tipos
    vectores = [vector.astype(np.float64) for vector in vectores]

    for i, vector in enumerate(vectores):
        c = len(vector)
        print(f"\nEl vector({i+1}) = {vector} tiene {c} componentes")
        try:
            if c != 3:
                raise ValueError("Error: Todos los vectores deben tener exactamente tres componentes para graficar en R3.")
        except ValueError as e:
            print(e)
            return

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Graficar los vectores
    for idx, vector in enumerate(vectores):
        ax.quiver(0, 0, 0, vector[0], vector[1], vector[2], color='C'+str(idx), label=f'Vector {idx+1}')

    # Calcular los límites automáticamente
    x_min = min(0, *np.min(vectores, axis=0))
    x_max = max(0, *np.max(vectores, axis=0))
    y_min = min(0, *np.min(vectores, axis=0))
    y_max = max(0, *np.max(vectores, axis=0))
    z_min = min(0, *np.min(vectores, axis=0))
    z_max = max(0, *np.max(vectores, axis=0))

    ax.set_xlim([x_min-1, x_max+1])
    ax.set_ylim([y_min-1, y_max+1])
    ax.set_zlim([z_min-1, z_max+1])

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.legend()

    plt.show()

# Ejemplo
vector1 = np.array([1, 2, 3])
vector2 = np.array([3, -1, 2])
vector3 = np.array([-2, 4, 1])

graficar_r3([vector1, vector2, vector3])
```

```python
#siendo r/ = resultado de variable
import numpy as np
import math

#con #p = sum(u*v) # sumamos los componentes del vector resultante de la multiplicacion
def angulo_entre_vectores(u,v,Num_ejercicio):
    #dos vectores son requeridos
    #hacemos el producto punto entre los vectores
    p = sum(u*v) #  r/ escalar resultante del producto punto

    #calculamos las normas de cada vector
    nu = np.linalg.norm(u)
    nv = np.linalg.norm(v)

    formula = p/(nu*nv) # r/ ...
    coseno_inverso = math.acos(formula) # arcocoseno

    # Convertir el ángulo de radianes a grados
    angulo_grados = math.degrees(coseno_inverso)

    if(angulo_grados == 180):
        tipo = "Llano"
    if(angulo_grados < 180 and angulo_grados > 90):
        tipo = "Obtuso"
    if(angulo_grados == 90):
        tipo = "Recto"
    if(angulo_grados == 360):
        tipo = "Completo"
    if(angulo_grados < 90 and angulo_grados > 0):
        tipo = "Agudo"
    if(angulo_grados < 360 and angulo_grados > 180):
        tipo = "Concavo"

    return print(f"\n\nAngulo entre vectores (ejercicio {Num_ejercicio}):\nAngulo en grados : {angulo_grados}°\nEs un angulo {tipo}\nAngulo en radianes : {coseno_inverso}")

#Ejemplo
#creacion de vectores / listas numericas
u = np.array([4,3,-1])
v = np.array([1,-1,1])

angulo_entre_vectores(u,v,34)
```

```python
#proyeccion de V sobre u proyuV
def proy_r2(u,v):
    m = np.sum(u*v)  # Producto punto (suma de cada componente)
    v_ = np.sum(u**2)  # Magnitud al cuadrado de U

    proy_u_V = (m/v_)*u  # Fórmula de la proyección

    print(f"\nLa proyeccion (vector3) de v sobre u resultante es \nProy = {proy_u_V}")
    # Graficar en R2
    graficar_r2([u, v, proy_u_V])


#vectores para el ejemplo
u = np.array([2,5])
v = np.array([3,-3])
proy_r2(u,v)
```
