```python
#first game

"""
Scatter plot game
Randomly generate points on a graph and the player has to input the (x,y) coordinates
For added difficulty, make the graph larger

"""
import random
import matplotlib.pyplot as plt

# Generar puntos aleatorios
def generar_puntos(num_puntos, rango):
    puntos = []
    for _ in range(num_puntos):
        x = random.randint(*rango)
        y = random.randint(*rango)
        puntos.append((x, y))
    return puntos

# Dibujar Scatter Plot
def dibujar_scatter_plot(puntos):
    xs = [p[0] for p in puntos]
    ys = [p[1] for p in puntos]
    plt.scatter(xs, ys)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Scatter Plot Game')
    plt.grid(True)
    plt.show()

# Juego principal
# Juego principal
def juego_scatter_plot(num_puntos, rango):
    puntos = generar_puntos(num_puntos, rango)
    dibujar_scatter_plot(puntos)

    # Interacción con el usuario
    intentos = 0
    while True:
        intentos += 1
        x_guess = int(input("Ingrese la coordenada X: "))
        y_guess = int(input("Ingrese la coordenada Y: "))

        # Verificar si las coordenadas están cerca de algún punto generado
        cerca_punto = False
        for punto in puntos:
            if abs(punto[0] - x_guess) < 5 and abs(punto[1] - y_guess) < 5:
                cerca_punto = True
                break

        if cerca_punto:
            print("Congratulation! te acercastes bien a un punto.")
            break
        else:
            print("Esas coordenadas no están cerca de ningún punto generado.")

    print("¡Juego completado en", intentos, "intentos!")


# Parámetros del juego
num_puntos = 20
rango = (0, 100)  # Rango de coordenadas (x, y)

# Ejecutar el juego
juego_scatter_plot(num_puntos, rango)
```

```python
#game two

"""
Algebra practice game
Generate one-step and two-step problems with random integer values and the player has to input the answer
Use positive and negative values. For added difficulty, make the numbers larger
"""
import random

def generate_problem():
    # Generar coeficientes aleatorios
    a = random.randint(-10, 10)
    b = random.randint(-10, 10)
    c = random.randint(-10, 10)

    # Generar la ecuación
    equation = f"{a}x + {b} = {c}"

    # Calcular la respuesta
    answer = (c - b) / a

    return equation, answer

def main():
    print("Bienvenido al juego de práctica de álgebra!")
    print("Resuelve la siguiente ecuación para x:")

    # Generar y mostrar el problema
    equation, correct_answer = generate_problem()
    print("Ecuación:", equation)

    # Obtener la respuesta del jugador
    player_answer = float(input("Ingresa tu respuesta para x: "))

    # Verificar la respuesta del jugador
    if player_answer == correct_answer:
        print("¡Respuesta correcta!")
    else:
        print("Respuesta incorrecta. La respuesta correcta es:", correct_answer)

if __name__ == "__main__":
    main()
```

```python
import random
import math

# Función principal del juego
def main():
    print("Welcome to Simple Projectile Game!")
    print("Try to clear the wall by adjusting the angle and velocity of the projectile.")

    # Generar altura y posición aleatoria del muro
    altura = random.randint(5, 15)
    posicion = random.randint(1, 20)

    # Solicitar al jugador que ingrese el ángulo y la velocidad
    angle = int(input("Enter the launch angle (0-90 degrees): "))
    velocity = int(input("Enter the velocity: "))

    # Calcular la distancia alcanzada por el proyectil
    distancia_alcanzada = velocity * math.cos(math.radians(angle)) * (velocity * math.sin(math.radians(angle)) + math.sqrt((velocity * math.sin(math.radians(angle)))**2 + 2 * 9.8 * altura)) / 9.8

    # Verificar si el proyectil supera el muro
    if distancia_alcanzada >= posicion:
        print("Congratulations! You cleared the wall!")
    else:
        print("Sorry, you didn't clear the wall. Try again!")

main()
```
